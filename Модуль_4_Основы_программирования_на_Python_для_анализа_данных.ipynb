**В системе аналитики DataFlow нужно написать функцию, которая будет считывать поток чисел (например, изменения показателей) до тех пор, пока их сумма не станет равна нулю.**

Когда баланс достигнут - функция должна вернуть сумму квадратов всех введённых чисел, чтобы аналитик мог оценить «силу колебаний» данных.

Вводные данные: Пользователь вводит целые числа по одному в строке.
Ввод завершается, когда сумма чисел становится равна 0.

2

2

-4

Выходные данные: Функция должна вывести одно число - сумму квадратов всех введённых значений.

24

```python
def sum_squares_sum():
  total_sum = 0
  sum_sq = 0
  while True:
    num = int(input())
    total_sum += num
    sum_sq += num ** 2
    if total_sum == 0:
      break
  return sum_sq

result = sum_squares_sum()
print(result)
```

```python
n = int(input()) #Считываем количество слов
words = [] #Создаем список для хранения слов
unique_counts = [] #Создаем список для хранения количества уникальных символов

for i in range(n):
  word = input() #считываем каждое слово
  words.append(word) #добавляем слово в список

for word in words:
  unique_chars = set(word.lower()) #создаем множество уникальных символов в нижнем регистре
  unique_counts.append(len(unique_chars)) #добавляем количество уникальных символов в список

for count in unique_counts:
  print(count) #вывод количества уникальных символов для каждого слова
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Загружаем встроенный датасет
california = fetch_california_housing()

# Создаём DataFrame из признаков
df_california = pd.DataFrame(california.data, columns=california.feature_names)
df_california
```

##**Задача №12**

Используя метод pd.concat([датафрейм1, датафрейм2], axis = 1), добавить к данным столбец, содержащий информацию о медианной стоимости дома (.target и .target_names).
В переменной california уже содержится(.target и .target_names):

```python
# Создаем DataFrame из целевой переменной
df_target = pd.DataFrame(california.target, columns=[california.target_names[0]])

# Объединяем два DataFrame
df_california = pd.concat([df_california, df_target], axis=1)

print("Датафрейм с добавленным столбцом медианной стоимости дома:")
print(df_california.head())
```
